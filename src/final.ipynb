```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Load your dataset
data = pd.read_csv('./dataset/score.csv')

# Generate a synthetic "Final Score"
np.random.seed(42)
data['Final Score'] = (
    0.4 * data['Soil Fertility Score'] +
    0.3 * data['Rainfall Score'] +
    0.3 * data['Greenscore']
) / data[['Soil Fertility Score', 'Rainfall Score', 'Greenscore']].max().sum() * 10

# Normalize features
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(data[['Soil Fertility Score', 'Rainfall Score', 'Greenscore']])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(normalized_features, data['Final Score'], test_size=0.2, random_state=42)

# Define the model
model = Sequential([
    Dense(64, activation='relu', input_shape=(3,)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='linear')
])

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=8,
    verbose=1
)

# Evaluate the model
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss:.4f}, Test MAE: {test_mae:.4f}")
```
